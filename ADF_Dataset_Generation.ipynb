```python
import pandas as pd
import random
from faker import Faker
import numpy as np
```

```python
fake = Faker()
num_records = random.randint(9000, 10000)  # Record count between 9k and 10k
```

```python
users = ["user1", "admin", "SYSTEM", "attacker", "guest"]
actions = ["ClearLogs", "InjectProcess", "AccountCreation", "DeleteUser", "AddService"]
files = ["config.sys", "log.txt", "data.db", "win.ini", "ntuser.dat"]
processes = ["svchost.exe", "cmd.exe", "explorer.exe", "powershell.exe", "malware.exe"]
persistence_techniques = ["ServiceInstall", "RunRegistry", "ScheduledTask", "None"]
ip_pool = ["10.0.0.5", "192.168.1.10", "172.16.0.2", None]
```

```python
data = []
for _ in range(num_records):
    row = {
        "Timestamp": fake.date_time_between(start_date="-5d", end_date="now").strftime("%m/%d/%Y %H:%M"),
        "User": random.choice(users),
        "Action": random.choice(actions),
        "File_Involved": random.choice(files),
        "Network_IP": random.choice(ip_pool),
        "Process_Name": random.choice(processes),
        "Registry_Modified": random.choice([0, 1]),
        "Persistence_Technique": random.choice(persistence_techniques),
        "Log_Cleared": random.choice([0, 1]),
        "Timestomped": random.choice([0, 1]),
        "Encoded_Payload": random.choice([0, 1]),
        "Encrypted_Channel": random.choice([0, 1]),
        "Uses_Anti_Forensics": random.choice([0, 1]),
    }
    data.append(row)
```

```python
df = pd.DataFrame(data)
df.to_csv("synthetic_adf_fp.csv", index=False)
```
